```python
# 기본 도구

import numpy as np
import pandas as pd

# 크롤링을 위한 도구

import requests
from bs4 import BeautifulSoup
import time

# 데베 연동을 위한 도구

import cx_Oracle
```

```python
url_string = "https://dhlottery.co.kr/gameResult.do?method=byWin&drwNo={0}"

winning_numbers = {}

for rnd in range(1, 11): #(1,920) : 1~919회차.
    winning_number = []
    response = requests.get(url_string.format(rnd))# html 가져오기
    soup = BeautifulSoup(response.text, 'html.parser') # BS : html의 데이터 꺼내올 수 있음
    spans = soup.select('.num.win span') #div class=num win
    for span in spans:
        winning_number.append(int(span.text))
    
    spans = soup.select('.num.bonus span')
    winning_number.append(int(spans[0].text))
    
    winning_numbers[rnd] = winning_number # winning_numbers[key] = value 구문. 정해진 key와 value가 들어간다.
    
    time.sleep(0.5) # 한 번 돌아갈 때마다 0.5초씩 쉬는걸루

    
winning_numbers
```

```python
conn = cx_Oracle.connect('hoseo', 'hoseo', '127.0.0.1:1521/xe')

cursor = conn.cursor()

cursor.execute("DELETE FROM LOTTO")

sql = """
     INSERT INTO LOTTO
     VALUES(:r, :n1, :n2, :n3, :n4, :n5, :n6, :b)
      """
for rnd, ns in winning_numbers.items(): # key,number가 들어옴
    cursor.execute(sql, r=rnd, n1=ns[0], n2=ns[1], n3=ns[2], n4=ns[3], n5=ns[4], n6=ns[5], b=ns[6])
    
conn.commit()
cursor.close()
conn.close()
```
