```python
import numpy as np
import analysis_helper as helper #helper함수 vscode에서 만들어 가져와서 씀
```

```python
region = '목4동'
result=helper.pop_ratio_by_region(region)
if result is not None:
    
    import csv
    import numpy as np
    

    total = []
    with open('data-files/population.csv', encoding='utf-8') as f:
        reader = csv.reader(f)
        lst = list(reader)
        lst = lst[1:] # header 빼기
        for row in lst:
            if region in row[0]:
                total.append(100) # 차이가 많이 나지 않아서 100을 추가
            else:
                result2 = np.array([ int(d.replace(',','')) for d in row[3:] ])
                result2 = result2 / np.sum(result2) # result2의 인구비율
                gap = result - result2 # result와 result2의 차이
                #total.append(np.sum(np.power(gap,2))) # 차이의 총량 (제곱)
                total.append(np.sum(np.abs(gap))) # 절대값
    #print(np.array(total))
    #print(np.argmax(total), np.argmin(total), np.max(total), np.min(total)) #argmax = max값이 어디에 있는지(위치값으로 나옴 [2]면 3번째)
    
    #print(lst[np.argmin(total)]) #양평동과 가장 비슷한 지역
    #print(lst[np.argmax(total)]) # 가장 차이나는 지역
    
    region1 = lst[np.argmax(total)] # my region (유사도 가장 높음)
    region2 = lst[np.argmin(total)] # another region
    
    region1 = [ int(x.replace(',','')) for x in region1[3:]] # ,를 공백으로 + 숫자화
    region2 = [ int(x.replace(',','')) for x in region2[3:]]
    
    import matplotlib.pyplot as plt
    plt.plot(region1, label= 'region 1', c='red')
    plt.plot(region2, label= 'region 2', c='green')
    plt.show()
   
    print(lst[np.argmin(total)]) #양평동과 가장 비슷한 지역
    print(lst[np.argmax(total)]) # 가장 차이나는 지역
    
else:
    print("검색 실패")
```
