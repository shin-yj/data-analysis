```python
X = np.linspace(-np.pi, np.pi, 256) # -3.14 ~ 3.14까지 256개로 균등 분할된 데이터로 반환
C = np.cos(X) # X에 대한 cos값 반환
S = np.sin(X) # X에 대한 sin값 반환
plt.plot(X, C, label='cos')
plt.plot(X, S, label='sin')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
plt.legend(loc=8)
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0) 
x2 = np.linspace(0.0, 2.0) 
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1) 
y2 = np.cos(2 * np.pi * x2)


ax1 = plt.subplot(1, 2, 1) # 1행 2열의 첫 번째 plot
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(1, 2, 2) # 1행 2열의 두 번째 plot
plt.plot(x2, y2, 'r.-') 
plt.xlabel('time (s)') 
plt.ylabel('Undamped')
plt.show()
```

```python
import csv
filtered_data = []
cnt = total = 0 # 정상 데이터의 정보 저장하는 변수

with open('data-files/seoul-temperature2.csv', encoding='utf-8') as hw:
    data = csv.reader(hw)
    next(data)
    # data[3] # 행
    for row in data: # 한 행 한 행 입력 -> row[2]면 여기선 열로 뽑힌다. 리스트에서 3번째 원소에 속하는 것을 뽑는 개념
        if row[4] != '': # 결측치(missing value)가 아닌 경우
            filtered_data.append(float(row[4]))
            total += float(row[4])
            cnt += 1
        else:
            filtered_data.append( -1000 ) # 나올 수 없는 값(-1000)을 넣어 결측치 처리함
            
# 전체 데이터 반복(range)하면서 결측치를 평균으로 대체
mean = total / cnt # 정상 데이터의 평균
for idx in range(len(filtered_data)) :
    if filtered_data[idx] == -1000:
        filtered_data[idx] = mean
```
