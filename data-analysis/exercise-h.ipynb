```python
# 위 작업은 탐색 연습
# 아래에는 일괄 처리 구현
```

```python
population_df = pd.read_csv('data-files/population.csv', encoding="utf-8", index_col=0)
population = population_df.copy()
population
```

```python
population = population.iloc[:, 2:]# 행은 다, 열은 2번째 부터
population.head()
```

```python
population['인구수합계'] = population.sum(axis=1)
population.head(2)
```

```python
population["인구수합계"]
```

```python
population.head(2)
```

```python
population_ratio = population.div(population["인구수합계"], axis= 0) # series를 행 방향 이름에 맞춰 계산해라, 행 방향(밑쪽)이 이름이 같으므로 # 리스트일 경우 이름이 없으니까 개수가 같아야 연산 가능하고 개수 다를경우는 다 NaN 갖는다.
population_ratio.head(2)
```

```python
my_region = population_ratio[population_ratio.index.str.contains('양평제1동')]
```

my_region

```python
my_region
```

```python
gap = population_ratio.sub(my_region.iloc[0], axis=1) # iloc[0]은 행만 이름있는 시리즈이므로 나이 기준으로 연산해야 한다 -> population_ ratio 에서 나이가 열방향에 있으므로 axis = 1
#gap = np.power(gap,2) # gap을 제곱해라 ( power는 n제곱 ) / np.abs는 절대값
gap = np.abs(gap) # power보다 값이 높아짐
summed_list = gap.sum(axis=1) # 열방향으로 sum
top_5_similar_regions = summed_list.sort_values(ascending=True)[:5] # 오름차순 (차가 제일 작은 순)
```

```python
my_region.iloc[0]
```

```python
top_5_similar_regions
```

```python
plt.rc('font', family='Malgun Gothic')
d = population_ratio.loc[top_5_similar_regions.index]
d = d.iloc[:, :-1]
d.T.plot(figsize=(20,10)) # 데이터 프레임을 그래프로 출력, 열 기준이므로  T를 빼면 0세~100세(열방향) 각각 그래프 그리게 됨. 따라서 가로방향인 T (Transpose)를 준다 

plt.show()
```
