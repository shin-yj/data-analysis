```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import scipy.stats
```

```python
np.random.seed(42) # seed값 같아야 같은 값  나옴

N = 10
mu_0 = 0.8 # 높은 값일수록 1이 많이 나옴

x = sp.stats.bernoulli(mu_0).rvs(N) # 평균이 0.5인 이항분포(0/1) random value selector n개
print("승-패 : ", x)

n= np.count_nonzero(x)
print("승리횟수 : [{0}] / 승률 : [{1}]".format(n, n/N))

p_value=sp.stats.binom_test(n, N) # 전체 시행횟수 중(N) true일 횟수(n) ->모집단의 귀무가설 mu=0.5(binom default)에서 뽑아 mu값이 0.8이 나올 확률 10% -> 0.5라고 생각하긴 이르다
p_value
```

```python
# obs = np.array([[5, 25], [10, 20]])
obs = np.array([[4, 16, 20], [23, 18, 19]])
print(obs)

chi2, p_value, dof, expected= sp.stats.chi2_contingency(obs)
print(chi2, dof, p_value) # 여기서 귀무가설은 관계 없다(=0)임. 관계 없을 확률 23% -> 유의수준 95%(5% 신뢰수준) 하에 귀무 기각할 수 없음 / 2. 모집단의 귀무가설은 관계가 없는데, 저 2x3의 데이터가 나올 확률 0.007 -> 모집단이 관계가 있다
```

```python
np.random.seed(0) 
x1 = sp.random.randint(70, 97, 20) # A학원 학생들의 수학 성적 
x2 = sp.random.randint(65, 90, 20) # B학원 학생들의 수학 성적
print(x1.mean(),x2.mean())
sp.stats.ttest_ind(x1,x2, equal_var=True) # 귀무 : 두 집단은 차이가 없다 -> x1,x2 12%의 확률 -> A학원 보낼 필요 X -> 귀무 채택 # t-test는 연속형 데이터만 가능(범주형 불가), 카이제곱은 범주형 데이터
# equal_var = 공분산
```

```python
N = 5
mu_1 = 0
mu_2 = 0.4 
np.random.seed(1) 
x1 = sp.stats.norm(mu_1).rvs(N) 
x2 = x1 + sp.stats.norm(mu_2, 0.1).rvs(N)

print(x1,x2)
print(x1.mean(), x2.mean())
print(sp.stats.ttest_ind(x1,x2, equal_var=False))
print(sp.stats.ttest_rel(x1,x2))
```
